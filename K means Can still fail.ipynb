```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=200,noise=0.10)
plt.style.use("s")
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
km.fit(x)
```

```python
centres = km.cluster_centers_
labels  = km.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=labels) 
plt.scatter(centres[:,0],centres[:,1],color="orange")
```

### Now to aviod this Kmeans clustering which is unable to separate the data points which are not Linearly seaparable we use DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbs  = DBSCAN(eps=0.2,min_samples=5)
```

```python
dbs.fit(x)
```

```python
ypred = dbs.fit_predict(x)
```

```python
print(ypred)
```

```python
plt.scatter(x[:,0],x[:,1],ypred,c='black')
```
